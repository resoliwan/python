```python
import matplotlib.pyplot as plt
import math
import normal
import seaborn
%matplotlib inline
```

```python
xs = [x / 10.0 for x in range(-50, 50)]
ys = [round(normal.normal_pdf(x),3) for x in xs]
print([(x, y) for (x,y) in zip(xs,ys) if x >= 0])
```

```python
plt.plot(xs, [normal.normal_pdf(x, mu=0, sigma=1) for x in xs], '-', label='mu=0, sigma=1')
plt.plot(xs, [normal.normal_pdf(x, mu=0, sigma=2) for x in xs], '--', label='mu=0, sigma=2')
plt.plot(xs, [normal.normal_pdf(x, mu=0, sigma=0.5) for x in xs], ':', label='mu=0, sigma=0.5')
plt.plot(xs, [normal.normal_pdf(x, mu=-1, sigma=1) for x in xs], '-.', label='mu=-1, sigam=1')
plt.legend()
plt.title('Various Noraml pdf')
plt.show()
```

```python
plt.plot(xs, [normal.normal_cdf(x, mu=0, sigma=1) for x in xs], '-', label='mu=0, sigma=1')
plt.plot(xs, [normal.normal_cdf(x, mu=0, sigma=2) for x in xs], '--', label='mu=0, sigma=2')
plt.plot(xs, [normal.normal_cdf(x, mu=0, sigma=0.5) for x in xs], ':', label='mu=0, sigma=0.5')
plt.plot(xs, [normal.normal_cdf(x, mu=-1, sigma=1) for x in xs], '-.', label='mu=-1, sigma=1')
plt.legend()
plt.show()
```
